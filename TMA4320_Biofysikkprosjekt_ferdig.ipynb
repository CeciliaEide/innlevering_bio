For å representere en polymer numerisk bruker vi et array av posisjonskoordinatene til hvert monomer. Det er et N x 2 array, der N er antall monomerer, og vi får N elementer av posisjonsvektorer på formen [x, y]. Indeksen i arrayet viser gjennomgående til et spesifikt monomers posisjon. Vi slipper da å lagre og vurdere alle de tomme posisjonene i koordinatsystemet og får en mer effektiv kode. Vi bruker et diskret koordinatsystem som er et kvadratisk gitter med avstand lik 1.

I flere av simuleringene tar vi utgangspunkt i en rett polymer og definerer derfor funksjonen makeStraightPolymer(N) som vist i kodeblokken under. Denne tar inn antall monomerer i polymeren, N, og gir ut en horisontal polymer midt i et koordinatsystem med størrelse N x N. Er N et partall vil den legge seg på den øverste av de to midterste radene, men dette påvirker ikke koden videre.

```python
def makeStraightPolymer(N): #Make a straight polymer with length N
    y = int(N/2) + 1 #arrange at 1/2 of length in y-direction (middle, depending on odd/even N)
    koord = np.zeros((N, 2), dtype=int)
    
    for i in range(N):
        koord[i] = [i+1, y]
    
    return koord #N x 2 matrix containing coordinates for monomer 1-N position, indexed 0 - (N-1)
```

En folding simuleres ved gjentatte rotasjoner av en ende av polymeret om en tilfeldig valgt monomer. Disse rotasjonene er forenklet ved at de foregår i et 2D-plan og at de alltid vil være en knekk på 90 grader, enten med eller mot klokka. Funksjonen rotatePolymer(polymer, koord, rot) tar inn en polymer, en tilfeldig rotasjonsretning og en tilfeldig monomer polymeret skal rotere om. Dette kommer vi tilbake til senere i prosjektet. Vi roterer alltid den korteste delen av polymeren slik at vi ikke havner utenfor koordinatsystemet vi har laget. Det blir ingen fysisk forskjell på å rotere den korte enden i en rotasjonsretning og på å rotere den lange enden i motsatt rotasjonsretning fordi sluttposisjonen blir lik, bare med annen orientering i rommet. Siden rotasjonsretning er tilfeldig kan vi dermed velge å alltid rotere den korte delen uten at det endrer resultatet. 

Vi ser også bort ifra rotasjoner om endemonomerene fordi det ikke gir noen energiendring. En eventuell energiendring for polymeren skyldes svake vekselvirkninger mellom monomerer som ikke allerede er bundet av kovalente bindinger. En rotasjon rundt et endepunkt på polymeren vil ikke gi en energiendring, da ingen monomerer endrer posisjoner relativt til hverandre og derfor ikke vil oppleve noen endrede vekselvirkninger.

Polymeren som system ønsker å minimere energien sin, og senere i oppgaven vil vi se at den som hovedregel kun endrer 
tilstand dersom den totale energiendringen er mindre enn null. Dersom en rotasjon om et endepunkt likevel inntreffer 
grunnet termiske fluktuasjoner som tillater ugunstige endringer, kan vi dessuten også se at dette ikke utgjør en reell 
endring for systemet vi simulerer. Vår løsning av oppgaven velger, som nevnt, alltid å rotere den korteste enden fra rotasjonsmonomeren, 
som for et endepunkt ikke vil inneholde noen monomerer i det hele tatt og derfor åpenbart ikke endre polymeren. Skulle vi 
likevel valgt å rotere den lengste enden, vil heller ikke monomerenes posisjoner endres relativt til hverandre. Å tillate 
en slik rotasjon i simuleringen vil være det samme som å endre polymerens orientering i koordinatsystemet, og utgjør ikke 
en reell forskjell for polymeren. Uansett metode vil rotasjon om endepunkt være et urelevant tilfelle, og vi har derfor 
vurdert at det ikke er hensiktsmessig å inkludere koordinater for endemonomerer i utvalget av rotasjonsmonomerer.

```python
def rotatePolymer(polymer, koord, rot): 
    polymer = np.copy(polymer)
    rot_om = np.where((polymer == koord).all(axis=1))[0][0]
    #for i in range(len(polymer)): #Kan ta (1, len(polymer) - 1) for raskere kode, siden vi ikke tar hensyn til rot om endemonomerer
        #if polymer[i][0] == koord[0] and polymer[i][1] == koord[1]:
            #rot_om = i
        
    if rot_om > len(polymer)/2: #Rotates monomers with index higher than the rotation-coordinate
        for i in range(rot_om + 1, len(polymer)):
            avstandsvektor = polymer[i] - polymer[rot_om]
            
            if rot == True: #Clockwise
                avstandsvektor[0], avstandsvektor[1] = avstandsvektor[1], -avstandsvektor[0]
                polymer[i] = polymer[rot_om] + avstandsvektor
            
            else: #Counter-clockwise
                avstandsvektor[0], avstandsvektor[1] = -avstandsvektor[1], avstandsvektor[0]
                polymer[i] = polymer[rot_om] + avstandsvektor
    
    else: #Rotates monomers with index lower than the rotation-coordinate
        for i in range(rot_om):
            avstandsvektor = polymer[rot_om] - polymer[i] 
            if rot == True: #Clockwise
                avstandsvektor[0], avstandsvektor[1] = avstandsvektor[1], avstandsvektor[0]
                polymer[i] = polymer[rot_om] + avstandsvektor
            
            else: #Counter-clockwise
                avstandsvektor[0], avstandsvektor[1] = -avstandsvektor[1], -avstandsvektor[0]
                polymer[i] = polymer[rot_om] + avstandsvektor
            
    return polymer
```

Vi har i rotasjonsjonsfunksjonen ikke tatt hensyn til om den nye polymeren er fysisk gyldig. En polymer skal være sammenhengende og den skal ikke foldes slik at monomerer havner i samme koordinater, slik at de ligger oppå hverandre. Vi har derfor laget funksjonen polymerIntact(polymer, N) som returnerer false dersom to eller flere monomerer ligger i samme koordinat eller om monomerer med kovalent binding, altså monomer med indeks [i] og monomer med indeks [i+1] eller [i-1], ikke er nærmeste nabo. Dette i henhold til oppgave 1 i prosjektbeskrivelsen.

Selv om ikke rotasjonsfunksjonen direkte kontrollerer at polymeren er sammenhengende gir den ikke noe rom for å lage en polymer som ikke er det. Vi har likevel valgt å inkludere en test for dette i polymerIntact for å best mulig besvare oppgaven.

```python
def polymerIntact(polymer, N): #Checking predefined rules for intact polymer (intact -> True)
    #Checking if polymer contains N monomers
    if len(polymer) != N:
        return False
    
    #Checking if any monomers are overlapping (two or more at same coord.)
    unique_positions = set()
    for monomer in polymer:
        # Convert numpy array to tuple for set insertion
        pos = tuple(monomer)
        unique_positions.add(pos)

    if len(unique_positions) != N:
        return False
    
    #Verifying that each monomer is neighbouring the previos one. (this should always be the case with the way our polymers are made and handeled throughut the project?)
    for i in range(len(polymer) - 1):
        dx = abs(polymer[i][0] - polymer[i + 1][0])
        dy = abs(polymer[i][1] - polymer[i + 1][1])
        if not ((dx == 1 and dy == 0) or (dx == 0 and dy == 1)):
            return False
    
    return True
```

Vi har nå satt opp forutsetninger for å simulere en velykket rotasjon av en polymer. Det er derimot mer interresant å se på en simulering med N_s forsøk på rotsasjon, noe vi har satt opp i funksjonen under. Denne tar inn en gyldig polymer, forsøker å folde den et gitt antall ganger (N_s) og returnerer sluttpolymererpolymeren og antall vellykkede rotasjoner. Vi gjør rotasjonsretning og hvilket monomer vi roterer om tilfeldig ved å generere pseudotilfeldige heltall fra numpy.random biblioteket. Denne modellen tar ikke hensyn til polymerens energi, noe vi kommer til senere.

```python
def N_simuleringer(N, N_s): #simulates N_s rotation attempts of a polymer of length N
    counter = 0
    polymer = makeStraightPolymer(N)
    
    for i in range(N_s):
        koord = polymer[np.random.randint(1, N-1)] #Excludes end-monomers
        rot = np.random.randint(0, 2)
        twisted_polymer = rotatePolymer(polymer, koord, rot)
        
        if polymerIntact(twisted_polymer, N) == True:
            counter += 1
            polymer = twisted_polymer
            
    return polymer, counter
```

Vi har nå beskrevet metoden vi skal bruke for å simulere en polymerfolding, men simuleringen for polymerfolding må ta høyde for at polymerene kan være lange og rotasjonsforsøkene mange. Vi har derfor utforsket hvordan vi kan skrive en kode med lik funksjonalitet, men som kjører på kortere tid. Vi har også oppdaget at måten vi behandler polymeren og dens rotasjoner i koordinatsystemet gjør det unødvendig å sjekke enkelte av kravene for en intakt polymer, da disse ikke kan bli brutt gjennom programmet. Vi har derfor også inkludert og analysert en versjon av koden som ekskluderer unødvendige sjekker. 

Nedenfor er "slow" den initielle koden, "fast" sikter til koden som gjør det samme men kjører raskere. "fastfast" er koden som er optimalisert og samtidig strippet for overflødige sjekker. Videre vil vi sammenligne og diskutere funksjonaliteten til de tre alternativene. 

```python
"""--------------------SLOW--------------------"""
def polymerIntact_s(polymer, N): #Checking predefined rules for intact polymer (intact -> True)
    #Checking if polymer contains N monomers
    if len(polymer) != N:
        return False
    
    #Verefies that monomers not overlap 
    for i in range(len(polymer)-1):
        for j in range(i+1, len(polymer)):
            result_variable = (polymer[i] == polymer[j]).all()
            if result_variable == True:
                return False
    
    #Checks if monomers are in a continuous chain
    test = 0

    for i in range(len(polymer)-1):
        if polymer[i+1][0] == polymer[i][0] and polymer[i+1][1] + 1 == polymer[i][1]:
            test += 1
        elif polymer[i+1][0] == polymer[i][0] and polymer[i+1][1] - 1 == polymer[i][1]:
            test += 1
        elif polymer[i+1][0] + 1 == polymer[i][0] and polymer[i+1][1] == polymer[i][1]:
            test += 1
        elif polymer[i+1][0] - 1 == polymer[i][0] and polymer[i+1][1] == polymer[i][1]:
            test += 1

    if test == len(polymer)-1:
        return True
    else:
        return False
    

def rotatePolymer_s(polymer, koord, rot): 
    polymer = np.copy(polymer)
    for i in range(len(polymer)): 
        if polymer[i][0] == koord[0] and polymer[i][1] == koord[1]:
            rot_om = i
        
    if rot_om > len(polymer)/2: #Rotates monomers with index higher than the rotation-coordinate
        for i in range(rot_om + 1, len(polymer)):
            avstandsvektor = polymer[i] - polymer[rot_om]
            
            if rot == True: #Clockwise
                avstandsvektor[0], avstandsvektor[1] = avstandsvektor[1], -avstandsvektor[0]
                polymer[i] = polymer[rot_om] + avstandsvektor
            
            else: #Counter-clockwise
                avstandsvektor[0], avstandsvektor[1] = -avstandsvektor[1], avstandsvektor[0]
                polymer[i] = polymer[rot_om] + avstandsvektor
    
    else: #Rotates monomers with index lower than the rotation-coordinate
        for i in range(rot_om):
            avstandsvektor = polymer[rot_om] - polymer[i] 
            if rot == True: #Clockwise
                avstandsvektor[0], avstandsvektor[1] = avstandsvektor[1], avstandsvektor[0]
                polymer[i] = polymer[rot_om] + avstandsvektor
            
            else: #Counter-clockwise
                avstandsvektor[0], avstandsvektor[1] = -avstandsvektor[1], -avstandsvektor[0]
                polymer[i] = polymer[rot_om] + avstandsvektor
            
    return polymer


def N_simuleringer_s(N, N_s): #simulates N_s rotation attempts of a polymer of length N
    counter = 0
    polymer = makeStraightPolymer(N)
    
    for i in range(N_s):
        koord = polymer[np.random.randint(1, N-1)] #Excludes end-monomers
        rot = np.random.randint(0, 2)
        twisted_polymer = rotatePolymer_s(polymer, koord, rot)
        
        if polymerIntact_s(twisted_polymer, N) == True:
            counter += 1
            polymer = twisted_polymer
            
    return polymer, counter
```

Merk: for enkelte kjøringer melder programmet om "division by zero"-problemer for de aller første stegene av koden nedenfor som sammenligner rask og treg funksjon. Feilen oppstår imidlertid kun på én av gruppemedlemmenes datamaskiner. Vi tolker dette som en avrundingsfeil for subtraksjon av vært like tall, og ønsker å oppgi en approksimasjon som sørger for at koden likevel kjører for den videre utregningen. Dette er ikke nødvendigvis en god tilnærming, og da feilen oppstår sjelden anbefales det å kjøre programmet igjen om feilen oppstår.

```python
d = 100 #number of datapoints

"""-------------------Tid som funksjon av N-------------------"""
N_array = np.linspace(5,100,d,dtype=int)

time_f = np.zeros(d)
time_s = np.zeros(d)
div_array = np.zeros(d)
count = 0

for N in N_array:
    start_f = time.time()
    poly_f = N_simuleringer_f(N, 100)[0]
    end_f = time.time()
    
    time_f[count] = end_f-start_f
    
    start_s = time.time()
    poly_s = N_simuleringer_s(N, 100)[0]
    end_s = time.time()
    
    poly_s = N_simuleringer_s(N, 100)[0]
    
    time_s[count] = end_s-start_s

    try:
        div_array[count] = (end_s-start_s) / (end_f-start_f)
    except:
        print("Error: float division by zero")
        div_array[count] = 0
    
    count +=1


plt.plot(N_array, time_f, 'g', label = "Fast function")
plt.plot(N_array, time_s, 'r', label = "Slow function")
plt.xlabel('Length of polymer, N')
plt.ylabel('Time per simulation set (N_s = 100) [sek]')
plt.title('Time per simultion as a function of N, fast vs. slow', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()

plt.plot(N_array, time_f, 'g', label = "Fast function")
plt.xlabel('Length of polymer, N')
plt.ylabel('Time per simulation set (N_s = 100) [sek]')
plt.title('Time per simultion as a function of N, fast', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()

plt.plot(N_array, div_array)
plt.xlabel('Length of polymer, N')
plt.ylabel('Relative speed (Slow/Fast)')
plt.title('Relative speed fast vs. slow as a function of N', fontsize=14)
plt.grid(True)
plt.show()
```

## Polymerens energi

Vi har til nå i prosjektet ikke tatt hensyn til vekselvirkninger mellom monomerene i polymerene, som kan gi opphav til en energiendring ved rotasjoner. Vi inkulderer dette i modellen ved å lage en vekselsvirkningsmatrise V, i funksjonen make_V(polymer, v_Energy), som angir vekselsvirkning mellom to monomerer. Indeksen V[i][j] viser vekselvirkningen mellom monomer i og j. Vi antar at monomerer kun vekselvirker hvis de er nærmeste naboer og at de ikke opplever vekselvirkning fra monomerene ved siden av seg i rekken, altså ikke mellom monomerer som har kovalent binding, eller fra seg selv. Derfor må elementene i diagonalen, over-, og underdiagonalen være lik null. Videre må matrisen være symmetrisk siden vekselvirkningen mellom monomer i og j må være lik vekselvirkningen mellom j og i (V[i][j] = V[j][i]). Når vi lager V i denne funksjonen antar vi at alle elementene som ikke er lik null har samme verdi, men dette er ikke egentlig noe krav. Vi vil se mer på dette senere i oppgaven.

```python
#making a matrix V 
def make_V(polymer, v_Energy):  
    N  = len(polymer)
    V = np.full((N, N), v_Energy)
    #zero elements in diagonal and upper- aand lower diagonal
    for i in range(N):
        for j in range(N):
            if i == j:
                V[i][j]=0
            elif i == j+1:
                V[i][j]=0
                V[j][i]=0
    return V
```

For å finne energien til et polymer må vi summere over alle vekselsvirkningene. Dette gjør vi med å multiplisere V-matrisen med operatoren b. b er en matrise med verdien 1 i koordinatene hvor vi har vekselsvirkende monomerer og null ellers. For eksempel vil vekselvirkning mellom monomer x og y gi veriden 1 i koorinatene (x-1, y-1) og (y-1, x-1) i matrisen b. Man finner da energien fra summen av elementene i den nye matrisen delt på to, slik at vekselvirkning mellom to monomerer ikke teller to ganger.

```python
#calculetes total energy of a polymer
def calculate_energy(polymer, V):
    polymer = np.copy(polymer)
    N = len(polymer)
    b = np.zeros((N, N))
    #making the b operator/E_matrix
    E_matrix = np.zeros((N, N))
    #inserting value 1 to the b operator, where there is interaction between mormones.
    for i in range(len(polymer)): 
        for j in range(i+2, len(polymer)):
            dx = abs(polymer[i][0] - polymer[j][0])
            dy = abs(polymer[i][1] - polymer[j][1])
            if ((dx == 1 and dy == 0) or (dx == 0 and dy == 1)):
                b[i][j] = 1
                b[j][i] = 1
    E_matrix = 1/2 * (V * b)
    E = np.sum((E_matrix))
    return E
```

For å simulere hvordan et polymer folder seg i en gitt temperatur, i en modell der vi også tar hensyn til energiendringen, 
bruker vi Metropolis algoritmen. Denne tar også utgangspunkt i tilfeldige rotasjoner, men roterer bare til posisjoner med 
høyere energi med en viss sannsynlighet. Dette gjør vi ved å bruke forholdet mellom sannsynligheten for at polymeren ligger i
de to ulike tilstandene. Forholdet er gitt ved $$e^{\beta \Delta E}$$ der beta er $$1/(k_b * T)$$ og deltaE er forskjellen i energi i 
de to tilstandene. Er energien til den roterte polymeren høyere enn den opprinnelige vil polymeren kun oppdateres dersom et tilfeldig generert tall mellom 0 og 1 er lavere enn dette forholdet, og har dermed simulert tilfeldige flukseringer til tilstander med større energi. Det er verdt å legge merke til at sannsynligheten for disse flukseringene er avhengig av temperaturen.

```python
#simulates polymers, where monomers are interacting, using Monte Carlo
def metropolisalgoritmen(polymer, N_s, V, T): 
    polymer = np.copy(polymer)
    beta = 1/(1.38 * 10**(-23) * T) #1/(k_b * T), [T] = K
    
    E_array = np.zeros(N_s + 1)
    E = calculate_energy(polymer, V)
    i = 0

    #twisting polymer and updates polymer if the new twisted polymer satisfies the "ruels" of polymerIntact(...)
    while i < N_s:
        koord = polymer[np.random.randint(1, len(polymer)-1)] #excluding the monomers at the ends
        rot = np.random.randint(0, 2)
        twisted_polymer = rotatePolymer(polymer, koord, rot)
        if polymerIntact(twisted_polymer, N):
            i += 1
            E_new = calculate_energy(twisted_polymer, V)
            if E_new < E:
                polymer = twisted_polymer
                E = E_new
            elif np.random.uniform(0, 1) < math.exp(-beta*(E_new - E)): 
                polymer = twisted_polymer
                E = E_new
            E_array[i] = E
    return polymer, E_array
```

En annen mulighet vi vil vurdere er å la en rett polymer foldes i synkende temperaturen, tilsvarende en avkjøling av polymeret. Vi ser nå på den lengste distansen over polymeret. Denne har sammenheng med energien da distansen blir lenger over en polymer som er mindre sammenkrøllet og dermed har høyere energi. Vi lager en funksjon maxDistance(polymer) som tar inn en polymer og returnerer den lengste distansen mellom to monomerer. Vi oppdaterer også Metropolisalgoritmen til å lagre alle de lengste distansene over polymeret som foldes i et nytt array. 

Deretter kjører vi to litt ulike funksjoner for å finne forventningverdien for den lengste distansen over polymeren, d, som funksjon av temperaturen. I den første starter vi med en rett polymer i hver av temperaturene og kutter de 1000 første stegene fordi disse ikke vil være representative for den gjeldene temperaturen. Vi starter i lave temperaturer og øker, men dette har egentlig ikke noe betydning for resultatet siden vi alltid starter med et rett polymer og forventningsverdien for hver temperatur blir uavhengig av hverandre. I den andre funksjonen starter vi i den høyeste temperaturen med et rett polymer og går deretter nedover i temperatur. Vi starter ikke nå med et rett molekyl for den lavere temperaturen, men bruker heller sluttpolymeren simulasjonen ved hakke høyere temperatur. Den rette polymeren er mer sannsynlig å finne ved høy temperatur og gir dermed en mer realistisk simulasjon.

```python
#finding the longest distance between two monomers in a polymer
def maxDistance(polymer):
    maxDistance = 1
    #iterating trough each monomer for each monomer and updates maxDistance
    for j in range(len(polymer)-1):
        for k in range(j+2, len(polymer)): #not checking the neighbor, since the distance is known and 1 
            d = np.linalg.norm(polymer[j] - polymer[k])
            if d > maxDistance:
                maxDistance = d
    return maxDistance
    
#same as the metropolisfunction, but using the maxDistance function to calculate the distance as well
def metropolisWithDistance(polymer, N_s, V, T): 
    polymer = np.copy(polymer)
    beta = 1/(1.38*10**(-23) * T) #1/(k_b * T), [T] = K
    
    E_array = np.zeros(N_s + 1) 
    d_array = np.zeros(N_s + 1)
    E = calculate_energy(polymer, V)
    d_array[0] = maxDistance(polymer)
    i = 0
    while i < N_s:
        koord = polymer[np.random.randint(1, len(polymer)-1)] 
        rot = np.random.randint(0, 2)
        twisted_polymer = rotatePolymer(polymer, koord, rot)
        if polymerIntact(twisted_polymer, N):
            i += 1
            E_new = calculate_energy(twisted_polymer, V)
            if E_new < E:
                polymer = twisted_polymer
                E = E_new

            elif np.random.uniform(0, 1) < math.exp(-beta*(E_new - E)): 
                polymer = twisted_polymer
                E = E_new
            E_array[i] = E
            d_array[i] = maxDistance(polymer) #making an array with distances so that we can calculate the mean value later
            
    return polymer, E_array, d_array 

#Calculates E[maxDistance] in the same way as for energy
def forventningsVal_d(polymer, N_s, V, T_arr):
    forventningsverdi = np.zeros(len(T_arr))
    
    for i in range(len(T_arr)):
        D_ar = metropolisWithDistance(polymer, N_s, V, T_arr[i])[2]
        D_ar = D_ar[1000:]
        forventningsverdi[i] = np.mean(D_ar)

    return forventningsverdi
#Calculates E[maxDistance] in the same way as for energy, but in the opposite temp. interval 
def forventningsVal_d_reversed(polymer, N_s, V, T_arr):
    forventningsverdi = np.zeros(len(T_arr))
    
    for i in range(len(T_arr)):
        polymer, E_var, D_ar = metropolisWithDistance(polymer, N_s, V, T_arr[i])
        D_ar = D_ar[1000:]
        forventningsverdi[i] = np.average(D_ar)

    return forventningsverdi

N = 20
N_s = 5000
T_list = np.arange(10, 1000, 50)
T_list_reversed = np.flip(T_list)

polymer_start = makeStraightPolymer(N)
vEnergy = np.random.randint(-6, -2) * 10**(-21)
V = make_V(polymer_start, vEnergy)
 
#timing
start = time.time()

forventningsverdi_d = forventningsVal_d(polymer_start, N_s, V, T_list)

end = time.time()
print("Execution time of the 1. program is- ", end-start, " seconds")

#timing
start = time.time()

forventningsverdi_d_reversed = forventningsVal_d_reversed(polymer_start, N_s, V, T_list_reversed)

end = time.time()
print("Execution time of the 2. program is- ", end-start, " seconds")

#Plotting E[maxDistance] and E[maxDistance] reversed as function of T

plt.xlabel('Temperatur [K]')  
plt.ylabel('Distanse') 
plt.title('Forventningsverdi av D som funksjon av T') 

plt.plot(T_list, forventningsverdi_d, label = 'Lav til høy')
plt.plot(T_list_reversed, forventningsverdi_d_reversed, label = 'Høy til lav')

plt.legend()
plt.show()
```
